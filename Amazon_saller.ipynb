```python
df['year'] = pd.DatetimeIndex(df['order_date']).year
df['date'] = pd.DatetimeIndex(df['order_date']).date
df['time'] = pd.DatetimeIndex(df['order_date']).time
df['month_name'] = pd.DatetimeIndex(df['order_date']).month_name()
df['day_name'] = pd.DatetimeIndex(df['order_date']).day_name()
df.drop('order_date', axis=1, inplace=True)
df.head(3)
```

```python
data=['quantity','item_total','shipping_fee']
ax = sns.boxplot(data=df[data], orient='h', palette='Set2')
```

```python
def ot(dt):
  sorted(dt)
  Q1 = dt.quantile(0.25)
  Q3 = dt.quantile(0.75)
  IQR = Q3 - Q1
  lower_range= Q1 - (1.5*IQR)
  upper_range= Q3 + (1.5*IQR)
  return lower_range, upper_range
```

```python
low,up = ot(df['item_total'])
df[(df['item_total']<low) | (df['item_total']>up)].index
```

```python
df['item_total'].fillna(df['item_total'].median(),inplace=True)
df['shipping_fee'].fillna(df['shipping_fee'].median(),inplace=True)
df['cod'].fillna('online', inplace=True)
df.head(10)
```

```python
days_sales=df[['year','month_name','day_name','item_total']].groupby(['year','month_name','day_name']).sum().reset_index()
```
